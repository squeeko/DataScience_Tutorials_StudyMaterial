```python
# Let's say there is a free lunch every 2nd Tuesday of the month. Find out those days
# for the entire year of 2017. 
```

```python
for mt in range(1,13):
    cal = calendar.monthcalendar(2017, mt)
    
    week_one = cal[0]
    week_two = cal[1]
    week_three = cal[2]
    
    if week_one[calendar.TUESDAY]:
        lunch_date = week_two[calendar.TUESDAY]
    else:
        lunch_date = week_three[calendar.TUESDAY]
    
    # print "{} : {}".format(mt, lunch_date)
    print "{} : {}".format(calendar.month_abbr[mt], lunch_date)
```
