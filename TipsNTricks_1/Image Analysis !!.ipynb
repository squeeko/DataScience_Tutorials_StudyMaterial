# Image Filtering

```python
# filtering - replacing a pixel value in an image with another pixel value
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import skimage

%matplotlib inline
```

```python
plt.rcParams['image.cmap'] = 'gray'
```

```python
normal = np.zeros(200)
normal[100:] = 15
```

```python
fig, axes = plt.subplots()
axes.plot(normal)
axes.margins(0.2)
```

```python
noise = normal + np.random.randn(*normal.shape)*2
```

```python
fig, axes = plt.subplots()
axes.plot(noise)
axes.margins(0.2)
```

```python
average = (noise[:-5] + noise[5:])/2
```

```python
fig, axes = plt.subplots()
axes.plot(average)
axes.margins(0.2)
```

# mean filter and gaussian filter

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import skimage

%matplotlib inline
```

```python
from skimage import filters
from scipy.ndimage import convolve
```

```python
square = np.zeros((10,10), dtype=float)
```

```python
print square
```

```python
square[3:7, 3:7] = 1
```

```python
print square
```

```python
fig, axes = plt.subplots()
axes.imshow(square)
```

```python
mean = np.ones((4,4))
```

```python
print mean
```

```python
mean = np.ones((4,4))/12.0
```

```python
print mean
```

```python
smooth_convolve = convolve(square, mean)
```

```python
smooth = filters.gaussian(square, 1.2)
```

```python
plt.imshow(square)
```

```python
plt.imshow(smooth_convolve)
```

```python
plt.imshow(smooth)
```
