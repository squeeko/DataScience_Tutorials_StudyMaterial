```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.violinplot(x=tips['tip'])
```

```python
tips.head()
```

```python
sns.violinplot(x=tips['total_bill'])
```

```python
sns.violinplot(x='size', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips) #smoker and sex
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', palette='husl')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='husl')
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, split=True)
```

```python
sns.violinplot(x='day', y='total_bill', hue='smoker', data=tips, split=True, palette='coolwarm')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='coolwarm', \
               order=["Sat", "Thur", "Sun", "Fri"])
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='coolwarm', \
               order=["Sat", "Thur", "Sun", "Fri"], hue='smoker')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, palette='RdBu', \
               order=["Sat", "Thur", "Sun", "Fri"], hue='smoker', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='coolwarm',\
              inner='quartile')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='coolwarm',\
              inner='quartile', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', palette='RdBu',\
              inner='quartile', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', palette='RdBu',\
              inner='quartile', split=True, scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', palette='RdBu',\
              inner='quartile', scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='husl',\
              inner='quartile', scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='husl',\
              inner='stick', scale='count')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='husl',\
              inner='stick')
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', palette='husl',\
              inner='stick', scale='count', split=True)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', palette='husl',\
              inner='stick', scale='count', scale_hue=False, split=True)
```
