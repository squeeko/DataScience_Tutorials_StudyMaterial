```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
# tips, iris
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips,color='green', kind='reg')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='hex')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='kde')
```

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=iris, kind='kde', color='red')
```

```python
# stat_func
```

```python
from scipy.stats import spearmanr
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, stat_func=spearmanr)
```

```python
sns.jointplot(x='size', y='total_bill', data=tips, stat_func=spearmanr, color='blue')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, ratio=4, size=6)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, ratio=7, size=6)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, ratio=2, size=5)
```
