```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
# iris, tips
```

```python
iris = sns.load_dataset('iris')
```

```python
sns.pairplot(iris)
```

```python
tips = sns.load_dataset('tips')
```

```python
sns.pairplot(tips)
```

```python
#hue
```

```python
sns.pairplot(iris, hue='species')
```

```python
#palette
```

```python
sns.pairplot(iris, hue='species', palette='husl')
```

```python
sns.pairplot(iris, hue='species', palette='RdBu')
```

```python
sns.pairplot(iris, hue='species', palette='Blues_d')
```

```python
# markers
```

```python
sns.pairplot(iris, hue='species', palette= 'husl', markers=['o', 'D', 's'])
```

```python
sns.pairplot(iris, hue='species', palette= 'husl', markers=['o', '+', 's'])
```

```python
# vars
```

```python
sns.pairplot(iris, vars=['sepal_length', 'sepal_width'])
```

```python
sns.pairplot(tips, vars=['total_bill','tip'])
```

```python
sns.pairplot(iris, vars=['sepal_length', 'sepal_width'])
```

```python
sns.pairplot(iris, size=4, vars=['sepal_length', 'sepal_width'])
```

```python
#x_vars, y_vars
```

```python
sns.pairplot(iris, x_vars=['petal_length', 'petal_width'], y_vars=['sepal_length', 'sepal_width'])
```

```python
sns.pairplot(iris, x_vars=['petal_length', 'petal_width'],\
             y_vars=['sepal_length', 'sepal_width'],\
            palette='husl', hue='species')
```

```python
sns.pairplot(iris)
```

```python
#diag_kind
```

```python
sns.pairplot(iris, diag_kind='kde')
```

```python
sns.pairplot(iris, diag_kind='kde', palette='husl', hue='species')
```

```python
sns.pairplot(iris, kind='reg')
```

```python
sns.pairplot(iris, diag_kind='kde', palette='husl', hue='species', kind='reg')
```

```python
sns.pairplot(tips, kind='reg')
```
