```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline 
```

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
flights = flights.pivot('month', 'year', 'passengers')
```

```python
sns.heatmap(flights)
```

```python
sns.clustermap(flights)
```

```python
sns.clustermap(flights, col_cluster=False)
```

```python
sns.clustermap(flights, row_cluster=False)
```

```python
sns.clustermap(flights, cmap='coolwarm', linewidths=1)
```

```python
sns.clustermap(flights, cmap='Blues_r', linewidths=1.4)
```

```python
sns.clustermap(flights, cmap='coolwarm', linewidths=1.2)
```

```python
sns.clustermap(flights, cmap='coolwarm', linewidths=1.2, figsize=(9,7))
```

```python
# standardize across col or rows
```

```python
sns.clustermap(flights, standard_scale = 1)
```

```python
sns.clustermap(flights, standard_scale = 0)
```

```python
# normalizing our dataset = z_score = 0 (rows)/1(col)
```

```python
sns.clustermap(flights, z_score=0)
```

```python
sns.clustermap(flights, z_score=1)
```
