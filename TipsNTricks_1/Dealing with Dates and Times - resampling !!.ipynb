```python
# Resampling is a common practice when dealing with dates
# changing a given data freq to a d/t freq
```

```python
import pandas as pd
import numpy as np
```

```python
date = pd.date_range('2016/05/05', periods = 55, freq = 'H')
```

```python
date_series = pd.Series(np.random.randn(len(date)), index=date)
```

```python
date_series
```

```python
date_series.shape
```

```python
# asfreq
```

```python
resample = date_series.asfreq('35Min', method='ffill')
```

```python
resample
```

```python
resample.shape
```

```python
resample2 = date_series.asfreq('120Min', method='ffill')
```

```python
resample2
```

```python
date_series.shape
```

```python
resample2.shape
```

```python
date_series.resample('3H').mean()
```

```python
date_series.resample('2H').sum()[1:20]
```

```python
date_series.resample('2H').mean()
```

```python
irregular_series = date_series[list(np.random.choice(a=list(range(len(date_series))),\
                                                     size=15, replace=False))]
```

```python
irregular_series
```

```python
irregular_series = irregular_series.sort_index()
```

```python
irregular_series
```

```python
irregular_series = irregular_series.asfreq('D')
```

```python
irregular_series
```

```python
irregular_series.resample('D').mean()
```

```python
irregular_series.resample('D').count()
```
